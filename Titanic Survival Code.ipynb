```python
data['Age']=data['Age'].fillna(data['Age'].mean())
data.isna().sum()
```

```python
#converting age column type from float to integer.
data['Age']=data['Age'].astype(int)
data
```

```python
sex_mapping = {"male": 0, "female": 1}
data['Sex'] = data['Sex'].map(sex_mapping)
data
```

```python
data.isna().sum()
```

```python
data['Cabin'].value_counts()
```

```python
data['Cabin'].fillna('No Cabin',inplace=True)
```

```python
data['Cabin'].value_counts()
```

```python
data['Has Cabin']=np.where(data['Cabin']=='No Cabin',0,1)
```

```python
data.head()
```

```python
data.drop(['Cabin'],axis=1,inplace=True)
data.drop(['PassengerId'],axis=1,inplace=True)
```

```python
data.head()
```

```python
mapping_dict={"S":0,"C":1,"Q":2}

data['Embarked']=data['Embarked'].map(mapping_dict)
```

```python
data.head(4)
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mean())
data.isna().sum()
```

```python
data['Name']
```

```python
data['Title'] = data['Name'].str.extract(r', (\w+\.)')
```

```python
data['Title'].value_counts()
```

```python
# Import label encoder
from sklearn import preprocessing
 
label_encoder = preprocessing.LabelEncoder()
 

data['Title']= label_encoder.fit_transform(data['Title'])
 
data['Title'].unique()
```

```python
data.head()
```

```python
data.drop(['Name','Ticket'],axis=1,inplace=True)
data.head(2)
```

```python
sns.pairplot(data,hue='Survived')
```

```python
X=data.drop(['Survived'],axis=1)
y=data['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()

lr.fit(X_train,y_train)
```

```python
y_pred_train = lr.predict(X_train)
print("y_pred_train =",y_pred_train)
```

```python
from sklearn import metrics
acc_train = metrics.accuracy_score(y_train,y_pred_train)
print("ACCURACY  = ",acc_train)
```

```python
y_pred_test = lr.predict(X_test)
print("y_pred_test =",y_pred_test)
```

```python
acc = metrics.accuracy_score(y_test,y_pred_test)
print("ACCURACY  = ",acc)
```
